```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
original = np.tile(np.linspace(50, 200, 256).astype(np.uint8), (256, 1))
```

```python
damaged = original.copy()
cv2.putText(damaged, 'TEXT', (60, 130), cv2.FONT_HERSHEY_SIMPLEX, 1.8, (255,), 4, cv2.LINE_AA)
cv2.rectangle(damaged, (180, 180), (240, 240), (0,), -1)

```

```python
mask = np.where(damaged != original, 255, 0).astype(np.uint8)
```

```python
image_float = damaged.astype(np.float32)
mask_bin = (mask > 0).astype(np.uint8)
```

```python
def pde_inpaint(img, mask, num_iter=500, dt=0.1):
    inpainted = img.copy()
    for _ in range(num_iter):
        lap = cv2.Laplacian(inpainted, cv2.CV_32F)
        inpainted += dt * lap * mask
    return np.clip(inpainted, 0, 255).astype(np.uint8)
```

```python
result = pde_inpaint(image_float, mask_bin, num_iter=500, dt=0.1)
```

```python
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1); plt.imshow(damaged, cmap='gray'); plt.title("Original (Damaged)")
plt.subplot(1, 3, 2); plt.imshow(mask, cmap='gray'); plt.title("Mask")
plt.subplot(1, 3, 3); plt.imshow(result, cmap='gray'); plt.title("Inpainted Result")
plt.show()
```

```python
cv2.imwrite("original.png", original)
cv2.imwrite("damaged.png", damaged)
cv2.imwrite("mask.png",mask)
cv2.imwrite("inpainted_result.png",result)
```
